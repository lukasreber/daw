```python
import pandas as pd
import numpy as np
import matplotlib as plt

```

```python
df = pd.read_csv('DAW_Exam_Material/tips.csv')
```

```python
df.info()
```

```python
df.day.unique()
```

```python
df['day'] = df['day'].astype('category')
df['day'] = df['day'].cat.set_categories(['Thur','Fri','Sat','Sun'], ordered=True)
df['day'].cat.categories
```

```python
# calculate tip percentage
df['percent_tip'] = 100/df['total_bill']*df['tip']
```

```python
df.groupby('day')['percent_tip'].mean().plot.bar()
```

```python
plt.style.use('seaborn')

df.boxplot('percent_tip',by='size')
```
